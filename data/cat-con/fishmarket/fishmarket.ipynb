```python
import pandas as pd
import warnings
import seaborn as sns
import matplotlib.pyplot as plt
from   pandas.plotting import radviz
sns.set(style="white", color_codes=True)
```

```python
fish_raw = pd.read_csv("fishmarket.csv")
fish_raw
```

```python
fish = fish_raw.drop(columns=["Length2", "Length3"])
```

```python
plt.figure(figsize=(15,10))
sns.pairplot(data=fish, hue="Species", height=4)
```

```python
plt.figure(figsize=(15,5))
sns.violinplot(data=fish, size=1, width=10, x="Species", y="Weight")
```

```python
plt.figure(figsize=(15,5))
sns.violinplot(data=fish, size=1, width=2, x="Species", y="Length1")
```

```python
plt.figure(figsize=(15,5))
sns.violinplot(data=fish, size=1, width=2, x="Species", y="Height")
```

```python
plt.figure(figsize=(15,5))
sns.violinplot(data=fish, size=1, width=2, x="Species", y="Width")
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(fish.corr(),annot=True,square=True)
```

```python
fish_melt = pd.melt(fish.drop(columns=["Weight"]), "Species", var_name="measurement")
fish_melt
```

```python
plt.figure(figsize=(15,10))
sns.swarmplot(data=fish_melt, x="measurement", y="value", hue="Species", size=10)
```

```python
plt.figure(figsize=(15,15))
radviz(fish, "Species")
```
